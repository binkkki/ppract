## Numpy

Типовой синтаксис для подключения numpy и matplotlib

```python
import numpy as np
from matplotlib import pyplot as plt
```

В отличие от списков массиви numpy позволяют реализовывать векторно-матричные операции обновременно над всеми элементами массива или матрицы

```python
# создаем массив чисел ль 1 до 10 с шагом 0.1
x = np.arange(0, 10, 0.1)
print(x)
y = np.sin(x)
print(y)
```

```python

t=np.arange(0, 360, 0.2)

x=4*np.cos(4*t) + 6*np.cos(5*t)
y=7*np.sin(1.5*t) + 5*np.sin(7*t)


plt.figure(figsize=(12,8))
plt.plot(x,y)
```
